#  Q 1.

```python
%matplotlib inline
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv('C:/Users/soura/Documents/STAT/AnovaQ1.csv')
#print(df)
df.boxplot()
df1=pd.melt(df.reset_index(), value_vars=['1','2','3'])
df1=df1.dropna()
#print(df1)
my_lm = ols('value ~ C(variable,Sum)',data=df1).fit()
print(my_lm.summary())
aov_table = sm.stats.anova_lm(my_lm,alpha=0.05)
print(aov_table)
from statsmodels.stats.multicomp import pairwise_tukeyhsd
m_comp = pairwise_tukeyhsd(endog=df1['value'], groups=df1['variable'], alpha=0.05)
print(m_comp)
```

# Boxplot Interpretation : 
As, we can see from the Boxplot column 1 w.r.t 2 seems to have different means as well as column 1 w.r.t 3 also seem to be different but column 2 w.r.t 3 seems to have less difference between their means so there is a high possibility for columns 2 and 3 to have different means making it to accept the Null Hypothesis.
# Tukey Test Result :
Now Multiple Comparison of Means using tukey test also states that group1=2 and group2=3 has their lower and upper values neagtive and positive ,i.e indicating 0 to lie in their interval due to which Null hypthesis cannot be rejected.
